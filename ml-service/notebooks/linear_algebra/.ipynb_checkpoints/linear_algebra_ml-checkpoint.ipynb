```python
# Скалярное произведение
dot_product = np.dot(book1_vector, book2_vector)
print("Скалярное произведение: ", dot_product)
```

```python
# Норма вектора (длина)
magnitude1 = np.linalg.norm(book1_vector)
magnitude2 = np.linalg.norm(book2_vector)
print("Норма первого вектора: ", magnitude1)
print("Норма второго вектора: ", magnitude2)
```

```python
# Косинусное сходство (cosine similarity)
cosine_sim = dot_product / (magnitude1 * magnitude2)
print("Косинусное сходство: ", cosine_sim)
```

```python
# Создание матрицы из векторов книг
books_matrix = np.array([book1_vector, book2_vector])
print("Матрица книг:\n", books_matrix)
print("Форма матрицы: ", books_matrix.shape)
```

```python
# Транспонирование матрицы
books_matrix_t = books_matrix.T
print("Транспонированная матрица:\n", books_matrix_t)
```

```python
# Матричное умножение (покажет совместимость признаков)
matrix_product = np.dot(books_matrix, books_matrix_t)
print("Результат умножения матриц:\n", matrix_product)
```
